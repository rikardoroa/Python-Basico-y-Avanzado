```python
#script by rikardoroa
#Outliers Detection
#Just python it!

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as snb
from matplotlib.colors import ListedColormap

class outlier_model():
    
   
    
    def __init__(self, df = pd.read_excel(r'D:\Contribuciones.xlsx',
                                          usecols=['Consumo_ContEle','id_contele']), df_final=pd.DataFrame()):
        #init
        self.df = df
        self.df_final = df_final
        
    def data_cleaning(self):
        #cleaning
        self.df.replace([np.inf, -np.inf], np.nan, inplace=True)
        self.df.dropna(inplace=True)
       
        
    def Interquartile_Calculation(self):
        #assign df
        df = self.df
      
        #calculating iq
        q1 = df['Consumo_ContEle'].quantile(0.25)
        q2 = df['Consumo_ContEle'].quantile(0.75)
        iqr = q2 - q1
        upper_bound = q2 + 1.5 * iqr
        lower_bound = q1 - 1.5 * iqr
        upper_filter = df['Consumo_ContEle']>upper_bound
        df3 = df[upper_filter]
        df3 = df3.assign(Data='Con Outlier')
        df3 = df3.assign(valor=1)
        lower_filter = df['Consumo_ContEle']<upper_bound
        df2 = df[lower_filter]
        df2 = df2.assign(Data='Sin Outlier')
        df2 = df2.assign(valor=0)
        
        #concatenating
        df4 = pd.concat([df2,df3], names=['Consumo_ContEle', 'id_contele'])
        
        #reseting index
        df5  = df4.reset_index(drop=True)
        self.df_final = df5
        
    def Outliers_Visualization(self):
        #init data
        df5 = self.df_final
        
        #Dimensions
        n_dims = (11.7, 8.27)
        
        # init figure and axes
        fig, ax = plt.subplots(figsize=n_dims)
        
        #hue args(outliers)
        hue_order = ['Sin Outlier','Con Outlier']
        
        #palette Colors
        colors = {'Sin Outlier': 'Blue', 'Con Outlier': 'red'}
        
        #scatter plot
        axes = snb.scatterplot(ax=ax,data=df5.sort_values('Data', key=np.vectorize(hue_order.index)), x='id_contele',
        y='Consumo_ContEle', hue='Data', hue_order=hue_order,palette=colors)
        
        #setting labels
        axes.set_xlabel('Id Consumo')
        axes.set_ylabel('Consumo Total en KwH')
        
        return plt.show()
    
    
    def Visualizacion_3D(self):
        df5 = self.df_final
        
        #setting style
        snb.set(style = "darkgrid")
        
        #setting dimensions and axes
        fig = plt.figure(figsize=(11.7, 8.27))
        ax = fig.add_subplot(111, projection = '3d')

        #setting Color
        valor = df5['valor']
        
        #labels
        labels=['Sin Outlier','Con Outlier']
        
        #settings X,Y 
        x = df5['id_contele']
        y = df5['Consumo_ContEle']
        z = df5['Consumo_ContEle']
        colors = ListedColormap(['blue', 'red'])
      
        
        #setting scatter and legend
        scatter = ax.scatter(x, y,z, c=valor, cmap=colors)
        ax.legend(handles=scatter.legend_elements()[0], labels=labels)

        #setting label axes
        ax.set_xlabel("Id Consumo")
        ax.set_ylabel("Consumo")

        #rotate de graph
        for angle in range(0, 360):
            ax.view_init(20, angle)
            plt.draw()
            
        #setting the grid to the graph
        ax.grid(True)
        return plt.show()


    
    
  
        
        
        
        
        
        
```

```python
df=outlier_model()
```

```python
df.data_cleaning()
```

```python
df.Interquartile_Calculation()
```

```python
df.Outliers_Visualization()
```

```python
df.Visualizacion_3D()
```
