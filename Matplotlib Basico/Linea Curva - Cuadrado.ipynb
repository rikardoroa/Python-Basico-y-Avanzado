```python
import matplotlib.pyplot as ptl
import numpy as np


class VisualizarData:

    def __init__(self, x = np.array(np.arange(60)), y = np.array([])):
        self.x = x
        self.y = y
        
    def calculo_datos(self):
        x = self.x
        y = self.y
        for index, item in np.ndenumerate(x):
            item = item ** 2
            y = np.insert(y, index, item).astype(int)
        self.y = y
        return y

    def visualiza_plot(self):
        x = self.x
        y = self.y
        figura = ptl.figure(figsize=(15, 15))
        ejes = figura.subplots()
        ejes.plot(x, y, '-gD', label='X^2')
        for xitem, yitem in np.nditer([x, y]):
            ejes.annotate(xitem, (xitem, yitem), textcoords="offset points", xytext=(0, 9), ha="center")
        ptl.legend(loc=2)
        ptl.grid(True)
        return ptl.show()
```

```python
visualizar = VisualizarData()
visualizar.calculo_datos()
visualizar.visualiza_plot()
```
