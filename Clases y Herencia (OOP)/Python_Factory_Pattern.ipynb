```python
#Python Factory design pattern
#Script by @rikardoroa


from abc import ABCMeta, abstractstaticmethod
    
    
class salary(metaclass=ABCMeta):
    
    @abstractstaticmethod #don't need to instanciate
    def get_sueldo():
        "this here is my interface from my python app"
        

#here comes the beauty of python
class Data_engineer_salary(salary):
    
    def __init__(self,name,age,role_salary):
        self.name = name
        self.age = age
        self.role_salary = role_salary
        
    
    
    def Data__engineer(self):
        role_salary = self.role_salary
        self.role_salary= 5000000 - (5000000 * 0.08)
        return self.role_salary
    
    
    
    def get_sueldo(self):
        return f'name:{self.name},age:{self.age},role_salary:{int(self.Data__engineer())}'
    
    
    
class Data_analytics_engineer_salary(salary):
    
    def __init__(self,name,age,role_salary):
        self.name = name
        self.age = age
        self.role_salary = role_salary
        
      
    def Data_analytics_engineer(self):
        self.role_salary= 3000000 - (3000000 * 0.08)
        return self.role_salary
    
    
    def get_sueldo(self):
        return f'name:{self.name},age:{self.age},role_salary:{int(self.Data_analytics_engineer())}'
    
    
    

class machine_learning_engineer_salary(salary):
    
    def __init__(self,name,age,role_salary):
        self.name = name
        self.age = age
        self.role_salary = role_salary
        
      
    def machine_learning_engineer(self):
        self.role_salary= 7000000 - (7000000 * 0.08)
        return self.role_salary
    
    
    def get_sueldo(self):
        return f'name:{self.name},age:{self.age},role_salary:{int(self.machine_learning_engineer())}'
    
    
    
class salaryFactory():
    
    @staticmethod
    def get_salary(salaryType):
        try: 
            if salaryType == "Data_engineer_salary":
                return Data_engineer_salary("Ricardo Roa",33,0)
            if salaryType == "Data_analytics_engineer_salary":
                return Data_analytics_engineer_salary("Luis pereira",30,0)
            if salaryType == "machine_learning_engineer_salary":
                return machine_learning_engineer_salary("Juan Guerra",33,0)
            
            raise AssertionError("Role not found!, please enter the correct rol!")
        
        except AssertionError as e:
            print(e)
            

if __name__ == "__main__":
    salary = salaryFactory.get_salary("Data_engineer_salary")
    print(f'{salary.get_sueldo()}')
    salary = salaryFactory.get_salary("Data_analytics_engineer_salary")
    print(f'{salary.get_sueldo()}')
    salary = salaryFactory.get_salary("machine_learning_engineer_salary")
    print(f'{salary.get_sueldo()}')

    
     
  
        
        
        
```
