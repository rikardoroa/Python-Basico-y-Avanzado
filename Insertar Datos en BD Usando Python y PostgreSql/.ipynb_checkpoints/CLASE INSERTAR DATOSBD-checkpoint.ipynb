```python
import pandas as pd
from sqlalchemy import create_engine




class insertar_datos_bd():
    
    def __init__(self, leerunidades=pd.DataFrame,leerobligaciones=pd.DataFrame, leerasignacion=pd.DataFrame, leerpagos=pd.DataFrame, 
                 leerasesor=pd.DataFrame, leertipologias=pd.DataFrame, leerdeudores = pd.DataFrame, leergestiones=pd.DataFrame):
        self.leerunidades = "C:\\Users\\Ricardo Roa\\Downloads\\PruebaTecnicaCoordinadorBI (2) (2)\\PruebaTecnicaCoordinadorBI\\recursos\\unidades.xlsx"
        self.leerobligaciones = "C:\\Users\\Ricardo Roa\\Downloads\\PruebaTecnicaCoordinadorBI (2) (2)\\PruebaTecnicaCoordinadorBI\\recursos\\obligaciones.xlsx"
        self.leerasignacion =  "C:\\Users\\Ricardo Roa\\Downloads\\PruebaTecnicaCoordinadorBI (2) (2)\\PruebaTecnicaCoordinadorBI\\recursos\\Asignacion.xlsx"
        self.leerpagos = "C:\\Users\\Ricardo Roa\\Downloads\\PruebaTecnicaCoordinadorBI (2) (2)\\PruebaTecnicaCoordinadorBI\\recursos\\pagos.csv"
        self.leerasesor = "C:\\Users\\Ricardo Roa\\Downloads\\PruebaTecnicaCoordinadorBI (2) (2)\\PruebaTecnicaCoordinadorBI\\recursos\\asesores.xlsx"
        self.leertipologias = "C:\\Users\\Ricardo Roa\\Downloads\\PruebaTecnicaCoordinadorBI (2) (2)\\PruebaTecnicaCoordinadorBI\\recursos\\tipologias.xlsx"
        self.leerdeudores = "C:\\Users\\Ricardo Roa\\Downloads\\PruebaTecnicaCoordinadorBI (2) (2)\\PruebaTecnicaCoordinadorBI\\recursos\\deudores.xlsx"
        self.leergestiones =  "C:\\Users\\Ricardo Roa\\Downloads\\PruebaTecnicaCoordinadorBI (2) (2)\\PruebaTecnicaCoordinadorBI\\recursos\\Gestiones.xlsx"
        
        
    def insertar_datos(self):
        leerunidades = pd.read_excel(self.leerunidades)
        insertar_datos = leerunidades[['id unidad','nombre','estado']]
        insertar_datos_unidad = insertar_datos.rename(columns={'id unidad':'id_unidad','nombre':'nombre_unidad'})
        self.leerunidades= insertar_datos_unidad
        return insertar_datos_unidad


    def insertar_obligacion(self):
        leerobligaciones = pd.read_excel(self.leerobligaciones)
        insertar_datos = leerobligaciones[['obligacion','estado']]
        insertar_datos_obligacion = insertar_datos.rename(columns={'obligacion':'id_obligacion'})
        self.leerobligaciones = insertar_datos_obligacion
        return insertar_datos_obligacion
    
    def insertar_asignacion(self):
        leerasignacion = pd.read_excel(self.leerasignacion)
        insertar_datos_asignacion = leerasignacion[['id','identificacion','obligacion','valor capital','valor actual','valor en mora',
                                                    'dias mora','unidad']]
        insertar_datos_asignacion = insertar_datos_asignacion.rename(columns={'id':'id_asignacion','obligacion':'id_obligacion','valor capital':'saldo_a_capital',
                               'valor actual':'saldo_actual','valor en mora':'saldo_en_mora','dias mora':'dias_mora'})
        
        self.leerasignacion = insertar_datos_asignacion
        return insertar_datos_asignacion
    
    
    def insertar_pagos(self):
        leerpagos = pd.read_csv(self.leerpagos , sep=';', quotechar='"', encoding = 'utf8')
        insertar_pagos = leerpagos[['obligacion','valor pago','fecha']]
        insertar_pagos = insertar_pagos.rename(columns={'obligacion':'id_obligacion','valor pago':'valor'})
        self.leerpagos = insertar_pagos
        return insertar_pagos
        
        
    def insertar_asesor(self):
        leerasesor = pd.read_excel(self.leerasesor)
        insertar_datos_asesor = leer_asesor[['asesor id','nombre','fecha ingreso','estado','unidad id']]
        insertar_datos_asesor = insertar_datos_asesor.rename(columns={'asesor id':'identificacion','nombre':'nombres','fecha ingreso':'fecha_ingreso',
                                                       'unidad id':'unidad'})
        self.leerasesor = insertar_datos_asesor
        return insertar_datos_asesor

    def insertar_tipologias(self):
        leertipologias = pd.read_excel(self.leertipologias)
        insertar_datos_tipologias = leertipologias[['cod','tipologia','contactabilidad','efectividad']]
        insertar_datos_tipologias = insertar_datos_tipologias.rename(columns={'cod':'id_tipologia'})
        self.leertipologias = insertar_datos_tipologias
        return insertar_datos_tipologias
    
    def insertar_deudores(self):
        leerdeudores = pd.read_excel(self.leerdeudores)
        insertar_deudores = leerdeudores
        self.leerdeudores = insertar_deudores
        return insertar_deudores

    
    def insertar_gestiones(self):
        leer_gestion = pd.read_excel(self.leergestiones)
        insertar_datos_gestion = leer_gestion[['idgestion','id asesor','fecha gestion','deudor','cod tipologia','hora gest']]
        insertar_datos_gestion = insertar_datos_gestion.rename(columns={'idgestion':'id_gestion','id asesor':'id_asesor','fecha gestion':'fecha_gestion',
                                                       'deudor':'identificacion_deudor','cod tipologia':'id_tipificacion','hora gest':'hora_gestion'})
        self.leergestiones = insertar_datos_gestion
        return insertar_datos_gestion

    

    def conexion_motor(self):
        insertar_datos_unidad=self.leerunidades
        insertar_datos_obligacion = self.leerobligaciones
        insertar_datos_asignacion = self.leerasignacion
        insertar_datos_pagos = self.leerpagos
        insertar_datos_asesor = self.leerasesor
        insertar_datos_tipologias = self.leertipologias
        insertar_datos_deudores = self.leerdeudores
        insertar_datos_gestion = self.leergestiones
        conexion_motor = create_engine('postgresql://postgres:milkas87@localhost:5432/PRUEBA')
        insertar_datos_unidad = insertar_datos_unidad.to_sql(
                'unidades', 
                conexion_motor,
                index=False, 
                if_exists='append')
        insertar_datos_obligacion= insertar_datos_obligacion.to_sql(
               'obligaciones', 
                conexion_motor,
                index=False, 
                if_exists='append')
        insertar_datos_asignacion.to_sql(
                'asignaciones', 
                conexion_motor,
                index=False, 
                if_exists='append'
                )
        insertar_datos_pagos.to_sql(
                'pagos', 
                conexion_motor,
                index=False, 
                if_exists='append'
                )
        insertar_datos_asesor.to_sql(
                'asesores', 
                conexion_motor,
                index=False, 
                if_exists='append'
                )
        insertar_datos_tipologias.to_sql(
               'tipologias', 
                conexion_motor,
                index=False, 
                if_exists='append'
                )
        insertar_datos_deudores.to_sql(
               'deudores', 
                conexion_motor,
                index=False, 
                if_exists='append'
                )
        insertar_datos_gestion.to_sql(
                'gestion', 
                conexion_motor,
                index=False, 
                if_exists='append'
                )
        return insertar_datos_unidad,insertar_datos_obligacion,insertar_datos_asignacion,insertar_datos_pagos, insertar_datos_asesor, insertar_datos_tipologias,insertar_datos_deudores, insertar_datos_gestion






```
