```python
# Script por rikardoroa
# image Scrapping with python
from urllib.request import urlopen
from urllib.request import urlretrieve
from bs4 import BeautifulSoup
import cv2
import pytesseract as pyt
from matplotlib import pyplot as plt

def ejecuta_tesseract():
    ejecuta_pyt = pyt.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'
    return ejecuta_pyt

def scrap_abrir_url():
    pag = urlopen('https://www.minsalud.gov.co/salud/publica/Vacunacion/Paginas/Vacunacion-covid-19.aspx')
    soup = BeautifulSoup(pag, "html.parser")
    img=soup.select('.seccion1chikun .interSec1Lchikun.left img')
    img_scr = img[0]
    pg_concatena = 'https://www.minsalud.gov.co/'+img_scr['src']
    pg_final = urlretrieve(pg_concatena, "vacunacion.jpg")
    image = cv2.imread('vacunacion.jpg')
    fig = plt.figure()
    fig.set_size_inches(15,15)
    image_show = plt.imshow(image)
    return image_show
```
