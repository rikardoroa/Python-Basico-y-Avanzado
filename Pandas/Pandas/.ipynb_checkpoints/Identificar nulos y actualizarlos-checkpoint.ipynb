```python
import pyodbc as con
import pandas as pd
import numpy as np


class filtroDatos:
    
    conn = con.connect(r'Driver={Microsoft Access Driver (*.mdb, *.accdb)};DBQ=C:\Users\Ricardo Roa\Documents\Database1.accdb;')
    conn2 = con.connect(r'Driver={Microsoft Access Driver (*.mdb, *.accdb)};DBQ=C:\Users\Ricardo Roa\Documents\Database2.accdb;')
                       
    def __init__(self, conn = conn,conn2 = conn,df = pd.DataFrame()):
        self.conn = conn
        self.conn2 = conn2
        self.df = df
    
    @classmethod
    def readQueryData(self):
        Db1 = pd.read_sql_query('''select * from dbo_CENSO010819''', self.conn)
        Db2 = pd.read_sql_query('''select * from dbo_CENSO020519''', self.conn2)
        Db_Data = pd.concat([Db2,Db1]).reset_index(drop=True)
        self.df = Db_Data
        return Db_Data.head()
    
    @classmethod
    def FillNullData(self):
        col = ["TRASLADO","PERMISO"]
        DbFinal_filtro = self.df.copy()
        for item in col:
            if (self.df[item].isnull()).any() == True:
                DbFinal_filtro[item] = np.where(self.df[item].isnull(), "NULL", self.df[item])
        return DbFinal_filtro.head()   
```
