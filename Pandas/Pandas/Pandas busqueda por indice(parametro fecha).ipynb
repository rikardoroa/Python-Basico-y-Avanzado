```python
import pandas as pd


class Vlookup():
    
    #declaro los dataframes como atributos de la clase
    SampleData = pd.read_excel(r"C:\Users\Ricardo Roa\Desktop\ejercicio cmodalidad.xlsx", 
                                        usecols=(["f_desembolso"]))
    OriginalData = pd.read_excel(r"C:\Users\Ricardo Roa\Downloads\TMX_Resoluc desde 2000.xlsx")
    LookupData = pd.DataFrame()
    
    #inicializando las variables
    def __init__(self, SampleData = SampleData , OriginalData = OriginalData,
                 LookupData = LookupData):
        
        self.SampleData = SampleData
        self.OriginalData =  OriginalData
        self.LookupData =  LookupData
        
    @classmethod
    def Formatdata(self):
        #invoco a los Dataframes Definidos en init
        SampleData  = self.SampleData 
        OriginalData =   self.OriginalData 
        
        #formateo las fecha a string
        SampleData[["f_desembolso"]] = SampleData["f_desembolso"].astype(str)
        OriginalData[["FechaVigencia"]] = OriginalData["FechaVigencia"].astype(str)
        
        #formateo los strings a fechas tipo datetime en ambos dataframes
        SampleData = pd.to_datetime(SampleData["f_desembolso"], format="%Y/%m/%d")
        OriginalDate = pd.to_datetime(OriginalData["FechaVigencia"], format="%Y/%m/%d")
        
        #le asigno la nueva columna(serie de datos) formateada al dataframe
        OriginalData[["FechaVigencia"]]  = OriginalDate
        
        #la data de muestra(fechas)  formateada la convierta en un dataframe
        SampleData = pd.DataFrame(SampleData)
        
        #capturo el dataframe en la variable inicializada
        self.SampleData = SampleData
        self.OriginalData = OriginalData
        return OriginalData.head(),SampleData.head()
    
    @classmethod       
    def Renamedata(self):
        #invoco el Dataframe final y renombro la serie
        FinaloriginalData = self.OriginalData
        FinaloriginalData = FinaloriginalData.rename(columns={"Variación anual porcentual % [Créditos en COP]":
                                                    "VariacionAPorcentualCCop"})
        
        
        self.OriginalData = FinaloriginalData #guardo la variable para su reutilizacion
        return FinaloriginalData.head()
    
    
    
    @classmethod  
    def VlookupData(self):
        #invoco las variables capturadas
        FinalOriginalData = self.OriginalData
        SampleData = self.SampleData
        new_res_data = self.LookupData
        
        #declaro una lista vacia para guardar las columnas
        cols = []
    
        #comienzo la iteracion de las columnas con el atributo columns 
        #del dataframe que contiene los datos de las resoluciones que en este caso es el dataFrame "FinalOriginalData"
        for col in FinalOriginalData.columns:
            cols.append(col)
        
        #declaro una lista para guardar los resultados del filtro aplicado al dataframe
        j = []
        
        #inicializo el contador en cero
        count = 0

        #iteracion con while para agregar las filas que coincidan con la condicion al dataframe
        while count < int(SampleData.shape[0]):
            #recorro la columna de la muestra de fechas
            for x in SampleData.f_desembolso:
                #recorro cualquier columna del dataframe "resoluciones" que llame como "FinalOriginalData"
                for i, y in FinalOriginalData.FechaVigencia.iteritems():
                    if x == y:#si los datos de la columnas son iguales
                        j.insert(count,FinalOriginalData.loc[i,cols])#inserte los datos
                        new_res = pd.DataFrame(j)#creo el dataframe
                count+=1 #hace un conteo por cada fila de la muestra (Dataframe fechas)
                
        return new_res #retorno el resultado
```
