```python
import json
import re


input_pattern1 = r"[0-9\s][\\/#+*.()${}%=?¡¿'´@A-Za-z\s]"
input_pattern2 = r"\s"


def write_json_credit_base():
    try:
        data = []
        index_data = []
        print("\n")
        print("importante!,si escoge esta opcion la base anterior se eliminara y se perderan todo los datos actualizados.")
        print("si ya tiene una base actualizada de creditos, se recomienda la opcion 2")
        print("pulse 0 para volver al menu")
        print("\n")
        json_input = input("digite la base de creditos que luego seran de consulta:")
        regex1 = re.search(input_pattern1,json_input)
        regex2 = re.search(input_pattern2,json_input)
        
        if json_input == '0':
            print("regresara al menu anterior")
            print("\n")
            json_option()

        elif bool(regex1) == True or bool(regex2) == True:
            print("error, digite el credito correctamente")

        else:
            data.append(json_input)
            for index, item in enumerate(data):
                new_item = item.split(",")
                list(new_item)

            for dict_index, item in enumerate(new_item):
                dict_index += 1
                index_data.append(dict_index)

            json_file = dict(zip(index_data,new_item))
            path = "C:\\Users\\Ricardo Roa\\Documents\\creditos.json"
            with open(path, 'w') as outfile:
                json.dump(json_file, outfile)
                
    except (FileNotFoundError):
            print("archivo no encontrado")

            
def update_json_file():
    try: 
        input_data=[]
        input_index_data=[]
        path = "C:\\Users\\Ricardo Roa\\Documents\\creditos.json"

        with open(path,'r') as f:
            json_data = json.load(f)


        first = len(json_data)
        input_item = input("agregue el nuevo credito a la base:")
        regex_input_pattern_one = re.search(input_pattern1,input_item)
        regex_input_pattern_two = re.search(input_pattern2,input_item)

        for key,item in json_data.items():
            if input_item == item:
                print(f"ya existe el credito {item} en la base")
        
        
        if bool(regex_input_pattern_one) == True or bool(regex_input_pattern_two) == True:
            print("error, digite el credito correctamente")
            json_option()

        else:

            input_data.append(input_item)
            for index, item in enumerate(input_data):
                new_data_item = item.split(",")
                list(new_data_item)



            for dict_index in range(len(new_data_item)):
                dict_index += 1
                dict_index = dict_index + first
                input_index_data.append(dict_index)


            json_new_data = dict(zip(input_index_data,new_data_item))
            json_data.update(json_new_data)


            path = "C:\\Users\\Ricardo Roa\\Documents\\creditos.json"
            with open(path, 'w') as outfile:
                json.dump(json_data, outfile)

            print("\n")
            print("base de creditos actualizada")
    
    except (FileNotFoundError):
            print("archivo no encontrado")

        
def json_option():
    while True:
        
        try:
            value = input("escoja una opcion del menu: 1\n"+ 
                          "1 : crear una nueva base para los creditos a liquidar\n"+
                          "2:  actualizar la base anterior de creditos con nuevos registros:")
            option = int(value)
            if option == 1:
                  return write_json_credit_base()
            if option == 2:
                  return update_json_file()
            else:
                print("opcion invalida, digite la opción correcta")
        except (ValueError, TypeError): #si hay error en la digitacion en digitar otro valor , genera nueva opcion
            print("error, opcion incorrecta, intente de nuevo")
```
