```python
#Script by RikardoRoa
#Just Python it
#Script Date = 19/09/2021

import numpy as np
import matplotlib.pyplot as ptl
import cmath
import math


class radians_graph:
    
    def __init__(self, complex_numbers=list, points = list):
        self.complex_numbers =  [3+2j,2+6j,3j]
        self.points  = points
        
    def calculate_radians(self):
        points = []
        for items in self.complex_numbers:
            x = cmath.phase(items)
            points.append(x)
        self.points = points
        return points 
    
    def sin_graph(self):
        points = self.points
        for rads in points:
            print(rads)
            x = np.arange(0,rads,0.025)
            y = np.sin(x)
            figura = ptl.figure(figsize=(24,9))
            ejes = figura.subplots()
            ejes.plot(x, y,  '-gD')
            titulo = ptl.suptitle('Representacion Grafica de algunos Radianes', fontname='arial', fontsize=20,fontweight='bold')
            ptl.setp(titulo,color='red')
            ptl.xlabel("Tiempo",fontsize=20)
            ptl.ylabel("Amplitud",fontsize=20)
            for xitem,yitem in np.nditer([x,y]):
                    etiqueta = "{:.2f}".format(xitem)
                    ejes.annotate(etiqueta, (xitem,yitem), textcoords="offset points",xytext=(0,10),ha="center")
                    ejes.grid(True)
        return ptl.show()
    
    
    def rad_todegrees(self):
        points = self.points
        degrees = []
        for rad in points:
            degree = math.degrees(rad)
            degrees.append(degree)
        return degrees
        
        

            
        
        
        
```

```python
points = radians_graph()
points.calculate_radians()
```

```python
points.sin_graph()
```

```python
points.rad_todegrees()
```
