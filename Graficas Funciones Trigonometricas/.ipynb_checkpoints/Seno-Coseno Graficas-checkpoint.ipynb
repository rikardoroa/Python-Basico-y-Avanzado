```python
"Script por @rikardoroa"
"Graficas correspondientes a seno y conseno utilizando los metodos Linspace ,Arange con Numpy y range"
"adicion de las etiquetas o labels con los metodos nditer y zip"
import numpy as np
import matplotlib.pyplot as ptl

def grafica_seno_coseno():
    x = np.arange(-4,2*np.pi, 0.3)
    y = 2*np.sin(x)
    y2 = 3*np.cos(x)
    figura = ptl.figure(figsize=(22,9))
    ejes = figura.subplots()
    ejes.plot(x, y,  '-gD')
    ejes.plot(x, y2, '-rD')
    titulo = ptl.suptitle('Grafica 2Sen(x) y 3Cos(x)', fontname='arial', fontsize=20,fontweight='bold')
    ptl.setp(titulo,color='red')
    ptl.xlabel("Tiempo",fontsize=20)
    ptl.ylabel("Amplitud",fontsize=20)
    for xitem,yitem in np.nditer([x,y]):
        etiqueta = "{:.1f}".format(xitem)
        ejes.annotate(etiqueta, (xitem,yitem), textcoords="offset points",xytext=(0,10),ha="center")
    for xitem,y2item in np.nditer([x,y2]):
        etiqueta2 = "{:.1f}".format(xitem)
        ejes.annotate(etiqueta2, (xitem,y2item), textcoords="offset points",xytext=(0,10),ha="center")
    ejes.grid(True)
    return ptl.show()
```

```python
grafica_seno_coseno()
```

```python
def grafica_seno_coseno3():
    x = np.arange(-4, 6, 0.2)
    y = (2*np.sin(x)) * ( 3*np.cos(x))
    figura = ptl.figure(figsize=(22,9))
    ejes = figura.subplots()
    ejes.plot(x, y,  '-gD')
    titulo = ptl.suptitle('Grafica 2Sen(x) X 3Cos(x)', fontname='arial', fontsize=20,fontweight='bold')
    ptl.setp(titulo,color='red')
    ptl.xlabel("Tiempo",fontsize=20)
    ptl.ylabel("Amplitud",fontsize=20)
    for x,y in zip(x,y):
        etiqueta = "{:.1f}".format(x)
        ejes.annotate(etiqueta, (x,y), textcoords="offset points",xytext=(0,10),ha="center")
    ejes.grid(True)
    return ptl.show()
```
